```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dp = pd.read_excel("AssignPlot1.xlsx")
x = dp['X']
y = dp['Y']
X = x.values.reshape(9,1)
Y = y.values.reshape(9,1)
plt.scatter(X,Y)
```

```python
def grad_decent(X,y,theta_val,learn_rate=0.01,iter=200):
    
    sample_c = len(y)
    
    cost_h = np.zeros(iter)
    theta_h = np.zeros((iter,2))
    
    for i in range(iter):
        prediction = np.dot(X,theta_val)
        
        theta_val = theta_val - (1/sample_c)*learn_rate*(X.T.dot((prediction-y)))
        theta_h[i,:] = theta_val.T
        
        cost = (1/2*sample_c) * np.sum(np.square(prediction-y))
        cost_h[i] = cost
        
    return theta_val,cost_h,theta_h
```

```python
learn_rate  =0.001
no_iterat = 800

theta = np.random.randn(2,1)
x_b = np.c_[np.ones((len(X),1)),X]
theta,cost_h,theta_h = grad_decent(x_b,Y,theta,learn_rate,no_iterat)
```
